## Acessando Matplotlib pelo m√©todo:

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
x = np.array(np.arange(10))
y = np.array(np.random.randn(10))
plt.plot(x, y, '-*', color='teal', linewidth=10, alpha=0.5)
plt.plot(y**2, x, color='red', linewidth=10)
plt.title('Titulo')
plt.ylabel('Eixo Y')
plt.xlabel('Eixo X')
```

```python
plt.subplot(1, 3, 1)
plt.plot(x, y, '--')
plt.title('Gr√°fico de pontilhado')
plt.subplot(1, 3, 2)
plt.plot(y, x, '*', color='pink')
plt.title('Star Graph')
plt.subplot(1, 3, 3)
plt.plot(x, y**2, '+', color='red')
plt.title('Cross Graph');
```

```python
fig = plt.figure()
ax = fig.add_axes([1, 1, 1, 1])
ax.plot(x, y , 'ro-')
plt.title('Titulo', fontweight='bold', style='italic', fontsize=25)
plt.xlabel('Eixo X', fontweight='bold')
plt.ylabel('Eixo Y', fontweight='bold')
ax2 = fig.add_axes([1.3, 1.6, 0.2, 0.3])
z = np.array(np.arange(11, 21))
ax2.plot(x, z**6,  label='Azul')
plt.title('Titulo 2')
plt.xlabel('Eixo X2')
plt.ylabel('Eixo Y2')
plt.legend(loc=0);
```

## Acessando Matplotlib pela classe:

```python
fig2, ax2 = plt.subplots(nrows=1, ncols=2)
ax2[0].plot(x, x**3, 'o-', color='green')
ax2[0].set_title('Titulo 1', fontweight='bold')
ax2[0].set_xlabel('Eixo X')
ax2[0].set_ylabel('Eixo Y')
ax2[1].plot(x**3, x, 'o-', color='yellow')
ax2[1].set_title('Titulo 2', fontweight='bold')
ax2[1].set_xlabel('Eixo X')
ax2[1].set_ylabel('Eixo Y')
plt.tight_layout()
```

```python
fig3, ax3 = plt.subplots()
ax3.plot(x, -z, 'ro-', label='Vermelho')
ax3.plot(-x, -z, '-+', color='teal', label='Azul Petroleo')
ax3.set_title('Pir√¢mide', fontsize=20, fontweight='bold')
ax3.legend();
```
