```python
import nltk
```

```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer
import pandas as pd
```

```python
sid = SentimentIntensityAnalyzer()
```

```python
df = pd.read_json('google_evandro.json')
df.replace(r'\n', '', regex=True, inplace=True)
```

```python
df
```

```python
df.isnull().sum()
```

```python
sid.polarity_scores(df.iloc[0]['Content'])
```

```python
df.Content[0]
```

```python
df['Scores'] = df['Content'].apply(lambda content: sid.polarity_scores(content))
```

```python
df
```

```python
df['Compound'] = df['Scores'].apply(lambda d: d['compound'])
```

```python
df['Overall Score'] = df['Compound'].apply(lambda score: 'POS' if score >=0 else 'NEG')
```

```python
df[df['Overall Score'] == 'NEG']
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer(max_df=.9, min_df=.1, stop_words= 'english')
```

```python
dtm = cv.fit_transform(df['Content'])
```

```python
dtm
```

```python
from sklearn.decomposition import LatentDirichletAllocation
```

```python
LDA = LatentDirichletAllocation(n_components=10)
```

```python
LDA.fit(dtm)
```

```python
len(cv.get_feature_names()) #Mesmo tamanho do dtm, ou seja, são mil e poucas dimensoes (documentos) por 590 linhas
```

```python
cv.get_feature_names()[100]
```

```python
len(LDA.components_) #Retorna o numero de tópicos
```

```python
first_topic = LDA.components_[0]
```

```python
type(first_topic)
```

```python
first_topic.argsort() #retorna um array com o numero dos indices dos valores sorteados em ordem do menor pro maior

#ex: [10, 200, 1] retornaria [2, 0, 1]
```

```python
top_ten = first_topic.argsort()[-10:]
```

```python
for index in top_ten:
    print(cv.get_feature_names_out()[index])
```

```python
import sys

orig_stdout = sys.stdout
f = open(
    'df.txt', 'w'
)
sys.stdout = f


for index, topic in enumerate(LDA.components_):
    dataframe = pd.DataFrame(columns=[f'THE TOP 15 WORDS FOR TOPIC #{index}'], data=[cv.get_feature_names_out()[index] for index in topic.argsort()[-15:]])
    print(dataframe)

sys.stdout = orig_stdout
f.close()
```

```python
df2 = pd.read_table('df.txt', header=None)
import numpy as np
arr = np.array(df2)
dataframe = pd.DataFrame(arr.reshape((10, 16))).T
headers = dataframe.iloc[0]
new_df = pd.DataFrame(data=dataframe.values[1:], columns=headers)
new_df
```

```python
topic_results = LDA.transform(dtm)
```

```python
topic_results[0].argmax()
```

```python
df['Topic'] = topic_results.argmax(axis=1)
```

```python
cont = pd.DataFrame(df['Contributor'].tolist(), columns=['col1', 'col2', 'col3', 'col4', 'col5'])
cont.replace('None', np.nan, inplace=True)
cont['Contributors'] = cont[['col1', 'col2','col3', 'col4', 'col5']].apply(lambda x: ', '.join(x[x.notnull()]), axis = 1)
role = pd.DataFrame(df['Role'].tolist(), columns=[
    'col1', 'col2', 'col3', 'col4', 'col5', 'col6', 'col7', 'col8', 'col9', 'col10',
    'col11', 'col12', 'col13', 'col14', 'col15', 'col16', 'col17', 'col18', 'col19', 'col20',
    'col21', 'col22', 'col23', 'col24', 'col25', 'col26', 'col27', 'col28', 'col29', 'col30',
    'col31', 'col32', 'col33', 'col34', 'col35', 'col36', 'col37', 'col38', 'col39', 'col40',
    'col41', 'col42', 'col43', 'col44'
])
role.replace('None', np.nan, inplace=True)
role['Roles'] = role[[
    'col1', 'col2', 'col3', 'col4', 'col5', 'col6', 'col7', 'col8', 'col9', 'col10',
    'col11', 'col12', 'col13', 'col14', 'col15', 'col16', 'col17', 'col18', 'col19', 'col20',
    'col21', 'col22', 'col23', 'col24', 'col25', 'col26', 'col27', 'col28', 'col29', 'col30',
    'col31', 'col32', 'col33', 'col34', 'col35', 'col36', 'col37', 'col38', 'col39', 'col40',
    'col41', 'col42', 'col43', 'col44'
]].apply(lambda x: ', '.join(x[x.notnull()]), axis = 1)
dff = pd.concat([df, cont['Contributors'], role['Roles']], axis=1)
dff.drop(['Contributor', 'Role'], axis=1, inplace=True)
dff = dff[['Title', 'Date', 'Contributors', 'Content', 'Link', 'Roles', 'Scores', 'Compound', 'Overall Score', 'Topic']]
dff
```

```python
positive = dff[dff['Overall Score'] == 'POS']
positive.reset_index(drop=True, inplace=True)
negative = dff[dff['Overall Score'] == 'NEG']
negative.reset_index(drop=True, inplace=True)
analytics = dff.groupby('Contributors').count().sort_values(by='Title', ascending=False)
analytics.to_excel('google_analytics.xlsx')
dff.to_excel('google_final.xlsx', index=False)
positive.to_excel('google_positive_articles.xlsx', index=False)
negative.to_excel('google_negative_articles.xlsx', index=False)
new_df.to_excel('google_topic.xlsx', index=False)
```
