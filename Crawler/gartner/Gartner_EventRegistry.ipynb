```python
import pandas as pd
import json
import re
```

```python
er = pd.read_json('event_registry.json')
er
```

```python
nested = pd.DataFrame(er.articles.values.tolist()).loc[2, 0]
frame = pd.DataFrame.from_records(nested)
event_registry = frame.drop(labels=['lang', 'isDuplicate', 'time', 'dateTime', 'dateTimePub', 'sim', 
                            'body', 'source', 'authors','image', 'eventUri', 'wgt'], axis=1)
event_registry_slice = event_registry.drop(['uri', 'dataType', 'url', 'sentiment', 'relevance'], axis=1)
event_registry_slice.columns = event_registry_slice.columns.str.replace('date', 'Date')
event_registry_slice.columns = event_registry_slice.columns.str.replace('title', 'Article')
event_registry_slice.Date = pd.to_datetime(event_registry_slice.Date)
event_registry_slice
```

```python
gartner = pd.read_csv('gartner.csv')
gartner.Date = pd.to_datetime(gartner.Date)
gartner
```

```python
df = pd.concat([event_registry_slice, gartner])
df.reset_index(drop=True, inplace=True)
df.to_csv('gartner_er.csv', index=False)
```
