```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from collections import Counter
from string import punctuation
from matplotlib import rcParams
import os
os.environ["KMP_DUPLICATE_LIB_OK"]="TRUE"
```

```python
import spacy
nlp = spacy.load("en_core_web_sm")
```

```python
df = pd.read_table('amazon.csv', sep=';')
df.replace(r'\n', '', regex=True, inplace=True)
df.replace(r'\xa0', '', regex=True, inplace=True)
df.replace(r'\r', '', regex=True, inplace=True)
df.replace(r'<h4>', '', inplace=True)
df.replace(r's:,', '', inplace=True)
df.replace(r'\n', '', inplace=True)
df.replace(r'\t', '', inplace=True)
df.replace(r'([01]?[0-9]|2[0-3]):[0-5][0-9]', '', regex=True, inplace=True)
df.replace(r':00', '', regex=True, inplace=True)
df.replace(r'UTC', '', regex=True, inplace=True)
df['Date'] = pd.to_datetime(df['Date'])
df.drop('Role', axis=1, inplace=True)
df
```

```python
def get_hotwords(text):
    result = []
    pos_tag = ['PROPN', 'ADJ', 'NOUN']
    doc = nlp(text.lower())
    for token in doc:
        
        if(token.text in nlp.Defaults.stop_words or token.text in punctuation):
            continue
        
        elif(token.pos_ in pos_tag):
            result.append(token.text)
                
    return result
```

```python
message_text = str(df.Title.values)
output = get_hotwords(message_text)
words = [x for x in Counter(output).most_common(10)]
```

```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer
sid = SentimentIntensityAnalyzer()
```

```python
df.dropna(inplace=True)
df.isnull().sum()
```

```python
sid.polarity_scores(df.iloc[0]['Content'])
```

```python
df.Content[0]
```

```python
df['Scores'] = df['Content'].apply(lambda content: sid.polarity_scores(content))
df['Compound'] = df['Scores'].apply(lambda d: d['compound'])
df['Overall Score'] = df['Compound'].apply(lambda score: 'POS' if score >=0 else 'NEG')
```

```python
df[df['Overall Score'] == 'NEG']
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_df=.9, min_df=.1, stop_words= 'english')
```

```python
dtm = cv.fit_transform(df['Content'])
dtm
```

```python
from sklearn.decomposition import LatentDirichletAllocation
LDA = LatentDirichletAllocation(n_components=10)
```

```python
LDA.fit(dtm)
```

```python
first_topic = LDA.components_[0]
```

```python
first_topic.argsort()
```

```python
top_ten = first_topic.argsort()[-10:]
```

```python
for index in top_ten:
    print(cv.get_feature_names_out()[index])
```

```python
import sys

orig_stdout = sys.stdout
f = open(
    'df.txt', 'w'
)
sys.stdout = f


for index, topic in enumerate(LDA.components_):
    dataframe = pd.DataFrame(columns=[f'THE TOP 15 WORDS FOR TOPIC #{index}'], data=[cv.get_feature_names_out()[index] for index in topic.argsort()[-15:]])
    print(dataframe)

sys.stdout = orig_stdout
f.close()
```

```python
dff2 = pd.read_table('df.txt', header=None)
import numpy as np
arr = np.array(dff2)
dataframe = pd.DataFrame(arr.reshape((10, 16))).T
headers = dataframe.iloc[0]
new_df = pd.DataFrame(data=dataframe.values[1:], columns=headers)
new_df
```

```python
topic_results = LDA.transform(dtm)
```

```python
df['Topic'] = topic_results.argmax(axis=1)
```

```python
positive = df[df['Overall Score'] == 'POS']
positive.reset_index(drop=True, inplace=True)
negative = df[df['Overall Score'] == 'NEG']
negative.reset_index(drop=True, inplace=True)
analytics = df.groupby('Contributor').count().sort_values(by='Title', ascending=False)
analytics.to_excel('alexa_analytics.xlsx')
df.to_excel('alexa_final.xlsx', index=False)
positive.to_excel('alexa_positive_articles.xlsx', index=False)
negative.to_excel('alexa_negative_articles.xlsx', index=False)
new_df.to_excel('alexa_topic.xlsx', index=False)
```
