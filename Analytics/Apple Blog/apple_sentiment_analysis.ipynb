```python
import pandas as pd
import nltk
from nltk.sentiment.vader import SentimentIntensityAnalyzer
```

```python
df = pd.read_json('apple2.json')
df.replace(r'\n', '', regex=True, inplace=True)
df.replace(r'\xa0', '', regex=True, inplace=True)
df.replace(r'\r', '', regex=True, inplace=True)
df.replace(r'<h4>', '', inplace=True)
df.replace(r's:,', '', inplace=True)
df.replace(r'\n', '', inplace=True)
df.replace(r'\t', '', inplace=True)
df.head()
```

```python
df.isnull().sum()
```

```python
sid = SentimentIntensityAnalyzer()
```

```python
sid.polarity_scores(df.iloc[0]['Content'])
```

```python
df['Scores'] = df['Content'].apply(lambda content: sid.polarity_scores(content))
df
```

```python
df['Compound'] = df['Scores'].apply(lambda d: d['compound'])
df['Overall Score'] = df['Compound'].apply(lambda score: 'POS' if score >=0 else 'NEG')
df.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_df=.9, min_df=.1, stop_words= 'english')
dtm = cv.fit_transform(df['Content'])
```

```python
dtm
```

```python
from sklearn.decomposition import LatentDirichletAllocation
LDA = LatentDirichletAllocation(n_components=10)
LDA.fit(dtm)
```

```python
first_topic = LDA.components_[0]
first_topic.argsort()
```

```python
top_ten = first_topic.argsort()[-10:]
for index in top_ten:
    print(cv.get_feature_names_out()[index])
```

```python
import sys

orig_stdout = sys.stdout
f = open(
    'apple_topic.txt', 'w'
)
sys.stdout = f


for index, topic in enumerate(LDA.components_):
    dataframe = pd.DataFrame(columns=[f'THE TOP 15 WORDS FOR TOPIC #{index}'], data=[cv.get_feature_names_out()[index] for index in topic.argsort()[-15:]])
    print(dataframe)

sys.stdout = orig_stdout
f.close()
```

```python
df2 = pd.read_table('apple_topic.txt', header=None)
import numpy as np
arr = np.array(df2)
dataframe = pd.DataFrame(arr.reshape((10, 16))).T
headers = dataframe.iloc[0]
new_df = pd.DataFrame(data=dataframe.values[1:], columns=headers)
new_df
```

```python
topic_results = LDA.transform(dtm)
topic_results[0].argmax()
```

```python
df['Topic'] = topic_results.argmax(axis=1)
df
```

```python
cont = pd.DataFrame(df['Contributor'].tolist(), columns=['col1', 'col2', 'col3', 'col4', 'col5', 'col6'])
cont.replace('None', np.nan, inplace=True)
cont['Contributors'] = cont[['col1', 'col2','col3', 'col4', 'col5', 'col6']].apply(lambda x: ', '.join(x[x.notnull()]), axis = 1)
dff = pd.concat([df, cont['Contributors']], axis=1)
dff.drop('Contributor', axis=1, inplace=True)
dff = dff[['Title', 'Date', 'Contributors', 'Content', 'link', 'Role', 'Scores', 'Compound', 'Overall Score', 'Topic']]
dff
```

```python
positive = dff[dff['Overall Score'] == 'POS']
positive.reset_index(drop=True, inplace=True)
negative = dff[dff['Overall Score'] == 'NEG']
negative.reset_index(drop=True, inplace=True)
analytics = dff.groupby('Contributors').count().sort_values(by='Title', ascending=False)
analytics.to_excel('apple_analytics.xlsx')
dff.to_excel('apple_final.xlsx', index=False)
positive.to_excel('apple_positive_articles.xlsx', index=False)
negative.to_excel('apple_negative_articles.xlsx', index=False)
new_df.to_excel('apple_topic.xlsx', index=False)
```
